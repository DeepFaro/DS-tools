### Adding RGB filter

```python
# import the necessary packages
import numpy as np
import cv2
import random
import pathlib
import matplotlib.pyplot as plt
```

```python
def random_rgb_red_filter(image):
    # split the image into its BGR components
    B, G, R = cv2.split(image)  
    
    r = random.randint(0,200)
    print(r)
    color = (0, 0, r)
    red_img  = np.full(image.shape, color, np.uint8)
    
    #fused_img = cv2.add(image, red_img)
    fused_img  = cv2.addWeighted(image, 0.8, red_img, 0.2, 0)

    return fused_img

def random_rgb_green_filter(image):
    # split the image into its BGR components
    B, G, R = cv2.split(image)  
    
    #r = random.randint(0,100)
    r = 255
    print(r)
    color = (0, r, 0)
    red_img  = np.full(image.shape, color, np.uint8)
    
    #fused_img = cv2.add(image, red_img)
    fused_img  = cv2.addWeighted(image, 0.7, red_img, 0.3, 0)
    
    return fused_img

def random_rgb_blue_filter(image):
    # split the image into its BGR components
    B, G, R = cv2.split(image)  
    
    r = random.randint(0,200)
    print(r)
    color = (r, 0, 0)
    red_img  = np.full(image.shape, color, np.uint8)
    
    #fused_img = cv2.add(image, red_img)
    fused_img  = cv2.addWeighted(image, 0.8, red_img, 0.2, 0)

    return fused_img

def random_rgb_all_filter(image):
    # split the image into its BGR components
    B, G, R = cv2.split(image)  
    
    color = (random.randint(0,255), random.randint(0,255), random.randint(0,255))
    red_img  = np.full(image.shape, color, np.uint8)
    
    #fused_img = cv2.add(image, red_img)
    fused_img  = cv2.addWeighted(image, 0.7, red_img, 0.3, 0)
    
    return fused_img
```

```python
image_path = r"C:\Users\ASUS\Documents\GitHub\DS-tools\cv_tools\data_aug_techniques\img\jn.jpg"
image = cv2.imread(image_path)
filtered1 = random_rgb_all_filter(image)
filtered2 = random_rgb_blue_filter(image)
filtered3 = random_rgb_green_filter(image)
filtered4 = random_rgb_red_filter(image)
#cv2.imshow("Images", np.hstack([image, filtered]))
#cv2.waitKey(0)
```
